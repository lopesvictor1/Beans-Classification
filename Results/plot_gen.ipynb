```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('exp_results.csv')

df
```

```python

sns.set_theme(rc={'figure.figsize':(12.7,10.27)})

ax = sns.barplot(df, y=df["Acurácia"], x=df.Nome, errorbar=None)
plt.xticks(rotation=90)
plt.title("Acurácia dos Experimentos")
ax.bar_label(ax.containers[0], fontsize=10);
ax.set(ylim=(80, 100))
ax.plot()
```

```python

sns.set_theme(rc={'figure.figsize':(12.7,10.27)})

ax = sns.barplot(df, y=df["Precisão"], x=df.Nome, errorbar=None)
plt.xticks(rotation=90)
plt.title("Precisão dos Experimentos")
ax.bar_label(ax.containers[0], fontsize=10);
ax.set(ylim=(80, 100))
ax.plot()
```

```python
#imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
import os
```

```python
#create the image folder

folder_path = os.path.join(os.getcwd(), 'Plots/Exploratory/')
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
#preparação dos dados

beans = fetch_ucirepo(id=602)
df = beans.data.features
df['Class'] = beans.data.targets
```

```python
#Distribuição das Classes


ax = sns.countplot(df, x='Class')
ax.bar_label(ax.containers[0], fontsize=10)
plt.title('Distribuição de Observações em Relação às Variedades de Feijão')
# TODO: mudar esse nome
plt.xlabel('Classes')
plt.ylabel('Frequência')
plt.savefig(os.path.join(folder_path, 'Distribuição de Observações em Relação às Classes.png'))

plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['Area'], bins=20, kde=True, color='blue')
plt.title('Distribuição da Área das Sementes de Feijão')
plt.xlabel('Área')
plt.ylabel('Frequência')
plt.savefig(os.path.join(folder_path, 'Distribuição da Área das Sementes de Feijão.png'))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
columns_to_include = df.columns[df.columns != 'Class']
sns.heatmap(df[columns_to_include].corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre Métricas')

plt.savefig(os.path.join(folder_path, 'Correlação entre Métricas.png'))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='Class', y='Area')
plt.title('Comparação da Área entre Variedades de Sementes de Feijão')
plt.xlabel('Variedade')
plt.ylabel('Área')
plt.savefig(os.path.join(folder_path, 'Comparação da Área entre Variedades de Sementes de Feijão.png'))
plt.show()
```
